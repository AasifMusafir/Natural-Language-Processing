**Lemmetization**

**WordNet Lemmetizers**

Lemmetization technique is like stemming. The Output we will get after lemmetization is called "lemma", which is a root word rather than a root stem, the output of stemming. After lemmatization,we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmattizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find lemma. Let us understand it with the example


USE CASES

1.Q&A

2.Chatbot

3.Text summarization

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')  # recommended for newer WordNet usage
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()

lemmatizer.lemmatize("going",pos="v")
```

**POS**

NOUN=n

VERB=v

Adjective=a

Adverb=r

```python
words=["eating","travelling","eaten","writing","writes","programming","historical","finally","finalized"]

for word in words:
    print(word+" -------> "+lemmatizer.lemmatize(word))
```

```python
print(lemmatizer.lemmatize("eating",pos="v"))
print(lemmatizer.lemmatize("eating",pos="a"))
print(lemmatizer.lemmatize("cleaning",pos="a"))
```
