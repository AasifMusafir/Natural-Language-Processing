**Stemming**

Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words know as lemma. Stemming is important in  natural language understanding(NLU) and natural language processing(NLP)

**PorterStemmer**

```python
from nltk.stem import PorterStemmer

stemming=PorterStemmer()

words=["eating","travelling","eaten","writing","writes","programming","historical","finally","finalized"]

for word in words:
    print(word+ " -----> " +stemming.stem(word))
```

```python
stemming.stem("Congratulations")
```

```python
stemming.stem("Kicking")
```

**RegexStemmer class(Regular Express Stemmer)**

NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer Algorithms. it basically takes a single regular expression and removes any prefix or suffix that matches the expression

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer("ing$|s$|able$|e$",min=4)

reg_stemmer.stem("Eat ing")
```

**Snowball Stemmer**

```python
from nltk.stem import SnowballStemmer

snow=SnowballStemmer("english")

for word in words:
    print(word+"----->"+snow.stem(word))
```

```python
#comparing Polar stemmer and Snowball stemmer
#Polar Stemmer
stemming.stem("sportingly"),stemming.stem("fairly")
```

```python
#Snowball stemmer
snow.stem("sportingly"),snow.stem("fairly")
```
